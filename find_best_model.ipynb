```python
import pathlib
import pandas as pd
import matplotlib.pyplot as plt

data_root = pathlib.Path() / 'run_data'
```

```python
result_files = list(data_root.glob('**/results.csv'))
results = {f"{f.parent.parent.parent.name}_{f.parent.name}": pd.read_csv(f) for f in result_files}
for data in results.values():
    data.columns = data.columns.str.strip()
```

```python
for name, data in results.items():
    if "epoch" in data.columns and "metrics/mAP50(B)" in data.columns:
        plt.plot("epoch", "metrics/mAP50(B)", data=data, label=name)
plt.legend()
plt.show()
```

```python
map_vals = [(k, data["metrics/mAP50(B)"].max()) for k, data in results.items()]
map_vals = sorted(map_vals, key=lambda x: x[1], reverse=True)
print(map_vals)
```
